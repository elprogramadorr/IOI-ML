Leer los datos

```python
import pandas as pd     
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
# eliminar columna Handle de Codeforces
df.drop('Handle de Codeforces', axis=1, inplace=True)
```

```python
print(len(df.loc[(df['Rating de Codeforces'].isna()) | (df['Rating de Codeforces'] == '')]) * 100 / len(df))

df = df.loc[(df['Rating de Codeforces'].notna()) & (df['Rating de Codeforces'] != '')]
```

```python
import matplotlib.pyplot as plt

# Datos
X = df['Rating de Codeforces']
Y = df['Ranking IOI 2024']

# Ajustar tamaño de la figura
plt.figure(figsize=(14, 8))

# Pintar las regiones invertidas (oro cerca del 100%)
plt.axvspan(90, 100, color='#FFEE55', alpha=0.3, label='Oro')   # Región Oro
plt.axvspan(70, 90, color='#CCCCCC', alpha=0.3, label='Plata')  # Región Plata
plt.axvspan(50, 70, color='#B87838', alpha=0.3, label='Bronce')  # Región Bronce

# Graficar los puntos
plt.scatter(Y, X, s=10, color='blue', label='Participantes')  # 's' controla el tamaño de los puntos

# Etiquetas y título
plt.xlabel('Ranking IOI 2024 (en %)')
plt.ylabel('Rating de Codeforces')
plt.title('Rating de Codeforces vs Ranking IOI 2024 por Regiones')

# Leyenda
plt.legend(loc='upper left')

# Ajustar márgenes
plt.tight_layout()
plt.xlim(0, 100)  # Ranking de 0% a 100% (porcentaje completo)
plt.ylim(min(X) - 50, max(X) + 50)  # Ajustar límites del Rating de Codeforces según los datos

# Mostrar el gráfico
plt.show()
```
